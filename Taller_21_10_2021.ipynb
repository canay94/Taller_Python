### ciclos anidados

```python
for i in range(0,3):
    for j in range(0,2):
        print(i, j)
```

```python
for i in range(0,3):
    for j in range(0,5):
        for k in range(0,6):
            print(i, j, k)
```

```python
for i in range(0,3):
    for j in range(0,5):
        for k in range(0,6):
        print(i, j, k)
```

```python
for i in range(0,1):
    for j in range(0,2):
        for k in range(0,3):
            print(i, j, k)
```
